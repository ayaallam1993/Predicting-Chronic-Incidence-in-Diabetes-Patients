# Check outliers:

```python
Med_df.columns
```

```python
# First, ensure the columns are numeric
Med_df['Avg_HBA1C Results'] = pd.to_numeric(Med_df['Avg_HBA1C Results'], errors='coerce')
Med_df['Age'] = pd.to_numeric(Med_df['Age'], errors='coerce')

# Now perform the IQR outlier detection
print("\nOutlier detection using IQR:")

numerical_cols_to_check = ['Avg_HBA1C Results', 'Age']

for col in numerical_cols_to_check:
    Q1 = Med_df[col].quantile(0.25)
    Q3 = Med_df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = Med_df[(Med_df[col] < lower_bound) | (Med_df[col] > upper_bound)]
    
    print(f"\n🔎 Column: {col}")
    print(f"Total outliers detected: {len(outliers)}")
    
    if not outliers.empty:
        print(outliers[[col]].head())
    else:
        print("No outliers detected.")
```
